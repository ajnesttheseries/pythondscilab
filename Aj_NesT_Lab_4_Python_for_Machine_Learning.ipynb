## 1. Mathematics: Calculation of Mean, Variance, and Standard Deviation in Statistics using Arrays

การทำ Machine Learning นั้นมีส่วนที่ใช้สถิติเข้ามาเกี่ยวข้องเสมอ ดังนั้นการเข้าใจสมการเบื้องต้นของสถิตินั้นเป็นสิ่งที่เข้าใจได้ ทั้งนี้เพื่อเป็นการทบทวนการใช้ Arrays จึงให้คำนวนสถิติโดยไม่ใช้ Library ดังนี้

- **การหาค่าเฉลี่ย (Mean)**
  *   ค่า Mean คือ ค่าเฉลี่ยที่เป็นการนำข้อมูลทั้งหมดมารวมกัน แล้วนำผลรวมดังกล่าวมาหารด้วยจำนวนของข้อมูลทั้งหมดที่นำมารวมกัน เช่น คะแนนผ่าน Mean หรือคะแนนไม่ผ่าน Mean หรือ สอบครั้งนี้ Mean อยู่ที่เท่าไหร่ 

-	**การหาค่าความแปรปรวน (Variance)**   
  *   ค่า Variance คือ ค่าความแปรปรวน เป็นการวัดการกระจายตัวของข้อมูล ที่นำผลรวมของความแตกต่างระหว่างข้อมูลแต่ละค่ากับค่า Mean มายกกำลัง 2 แล้วหารด้วยจำนวนข้อมูล
  *   โดยค่า Variance ที่มีค่ามากจะแสดงให้เห็นถึงการกระจายของข้อมูลที่ห่างกันมาก ในขณะที่ค่า Variance ที่มีค่าน้อยก็แสดงถึงการกระจายของข้อมูลที่ใกล้เคียงกันมากขึ้น
- **การหาค่าส่วนเบี่ยงเบนมาตรฐาน (Standard Deviation)**
  *   ค่า Standard Deviation คือ ค่าส่วนเบี่ยงเบนมาตรฐาน เป็นการวัดการกระจายของกลุ่มข้อมูล ที่นำค่า Variance มา Square Root ซึ่งในการนำไปใช้ส่วนใหญ่จะใช้ดูคู่กับค่า Mean เช่น ผลการสอบ 2 กลุ่มได้ค่า Mean เท่ากัน ถ้าจะแปลผลว่า 2 กลุ่มนี้เก่งเท่ากันจะไม่ถูกต้อง ต้องดูค่า S.D. ของกลุ่มไหนน้อยกว่า แสดงว่ากลุ่มนั้นเก่งกว่าถึงแม้ค่า Mean จะเท่ากัน 
  *   ค่า S.D. มีค่าน้อยแสดงว่าข้อมูลมีการกระจายที่ใกล้เคียงกันมากกับค่าเฉลี่ย แต่หากค่า S.D. มีค่ามากแสดงว่าข้อมูลมีการกระจายที่ห่างกันมากกับค่าเฉลี่ย

````python
"""
Aj. NesT the Series
1. Mathematics: Calculation of Mean, Variance, and Standard Deviation in Statistics using Arrays
Exercise 4.1
"""
x = [40, 170, 95, 35, 45, 40, 125, 85, 30, 75]    # ความยาว (มม.)
y = [20, 9, 7, 17, 15, 35, 8, 7, 12, 10]          # ความกว้าง (มม.)
z = ['R','P','P','R','R','R','P','P','R','P']     # วัตถุ (R = ยางลบ, P = ดินสอ)
xAvg, yAvg = 0, 0

# วนบวกค่าทั้งใน Array เข้าด้วยกัน
for i in range(len(x)):
  xAvg += x[i]
  yAvg += y[i]    

# หารค่าที่รวมมาทั้งหมดด้วยจำนวนข้อมูล
xAvg = xAvg/len(x)
yAvg = yAvg/len(y)
print("Average of items length: " + str(xAvg))
print("Average of items width: " + str(yAvg))
````

````python
x = [40, 170, 95, 35, 45, 40, 125, 85, 30, 75]    # ความยาว (มม.)
y = [20, 9, 7, 17, 15, 35, 8, 7, 12, 10]          # ความกว้าง (มม.)
z = ['R','P','P','R','R','R','P','P','R','P']     # วัตถุ (R = ยางลบ, P = ดินสอ)
xAvg, yAvg = 0, 0
RxAvg, RyAvg = 0, 0
PxAvg, PyAvg = 0, 0
RCount, PCount = 0, 0

# วนบวกค่าทั้งใน Array เข้าด้วยกัน
for i in range(len(x)):
  xAvg += x[i]
  yAvg += y[i]    
  if z[i] == "R":
      RxAvg += x[i]
      RyAvg += y[i]
      RCount += 1
  elif z[i] == "P":
      PxAvg += x[i]
      PyAvg += y[i] 
      PCount += 1   

# หารค่าที่รวมมาทั้งหมดด้วยจำนวนข้อมูล
xAvg = xAvg/len(x)
yAvg = yAvg/len(y)
RxAvg = RxAvg / RCount
RyAvg = RyAvg / RCount
PxAvg = PxAvg / PCount
PyAvg = PyAvg / PCount

print("Items")
print("Average of items length: " + str(xAvg))
print("Average of items width: " + str(yAvg))

print("\nRubber")
print("Average of rubber length: " + str(RxAvg))
print("Average of rubber width: " + str(RyAvg))

print("\nPencil")
print("Average of pencil length: " + str(PxAvg))
print("Average of pencil width: " + str(PyAvg))
````

**การหาค่าความแปรปรวนหรือการกระจายตัว (Variance)**
	
ค่าความแปรปรวนหรือการกระจายตัว คือ ค่าที่แสดงถึงการกระจายของข้อมูล เช่น [1,2,3] ซึ่งจะได้ค่าเฉลี่ยคือ 2 (mean)
จาก (1 + 2 + 3) / 3) = 2 

จากนั้นนำค่าเฉลี่ยที่ได้นำมาลบกับข้อมูลที่มีจากนั้นนำมายกกำลัง 2 และนำมารวมกันหาค่าเฉลี่ยจะทำให้ได้ค่าการกระจายตัวของข้อมูลซึ่งจะสามารถอธิบายได้ดังรูปด้านล่าง

จากตารางข้างบนจะได้ค่าการกระจายตัวดังนี้ (1 + 0 + 1) / 3 = 0.67 (คอลัมน์ขวาสุดรวมกันแล้วหารด้วยจำนวนแถวที่มี)

**การหาค่าส่วนเบี่ยงเบนมาตรฐาน (Standard Deviation)**

ใช้สำหรับการวัดค่าการกระจายตัวของข้อมูลที่ออกไปจากค่าเฉลี่ย (Mean) สามารถทำได้โดยการทำค่าการกระจายตัว  (Variance ) มา Square Root จะได้ค่าดังนี้

image.png

image.png

การหาค่าการกระจายตัว (Variance) และส่วนเบี่ยงเบนมาตรฐาน (Standard Deviation) โดยไม่ใช้ numpy

````python
import math 
value, Buffer = [1,2,3], [0,0,0]  # ประกาศ 2 array พร้อมกัน
Average, Variance, SD = 0,0,0     # ประกาศ paramter 3 ตัวพร้อมกัน
# Mean
for i in range(len(value)):
  Average += value[i]
Average /= len(value)

# Variance
for i in range(len(value)):
  Buffer[i] = (value[i] - Average)**2
  Variance += Buffer[i]
Variance /= len(value)

# Standard Deviation
SD = math.sqrt(Variance)

# Display
print("Average: " + str(Average))
print("Variance: " + str(Variance))
print("Standard Deviation: " + str(SD))
````

ใช้ Numpy

**ฝึกปฏิบัติการเขียนโปรแกรม**

````python
import math 
RubberL, Buffer_1 = [40,95,35,45,30], [0,0,0,0,0]
RubberW, Buffer_2 = [20,17,15,35,12], [0,0,0,0,0]
Average_1, Variance_1, SD_1 = 0,0,0
Average_2, Variance_2, SD_2 = 0,0,0

PencilL, Buffer_3 = [170,95,125,85,75], [0,0,0,0,0]
PencilW, Buffer_4 = [9,7,8,7,170], [0,0,0,0,0]
Average_3, Variance_3, SD_3 = 0,0,0
Average_4, Variance_4, SD_4 = 0,0,0

# หาค่าเฉลี่ย
for i in range(len(RubberL)):
    Average_1 += RubberL[i]
    Average_2 += RubberW[i]
    Average_3 += PencilL[i]
    Average_4 += PencilW[i]
Average_1 /= len(RubberL)
Average_2 /= len(RubberW)
Average_3 /= len(PencilL)
Average_4 /= len(PencilW)

# หาค่าความแปรปรวน
for i in range(len(RubberL)):
    Buffer_1[i] = (RubberL[i] - Average_1) ** 2
    Variance_1 += Buffer_1[i]
    Buffer_2[i] = (RubberW[i] - Average_2) ** 2
    Variance_2 += Buffer_2[i]
    Buffer_3[i] = (PencilL[i] - Average_3) ** 2
    Variance_3 += Buffer_3[i]
    Buffer_4[i] = (PencilW[i] - Average_4) ** 2
    Variance_4 += Buffer_4[i]
Variance_1 /= len(RubberL) 
Variance_2 /= len(RubberW) 
Variance_3 /= len(PencilL)
Variance_4 /= len(PencilW)

# หาค่าส่วนเบี่ยงเบนมาตรฐาน
SD_1 = math.sqrt(Variance_1)
SD_2 = math.sqrt(Variance_2)
SD_3 = math.sqrt(Variance_3)
SD_4 = math.sqrt(Variance_4)

# แสดงผล
print("Rubber Length")
print("Average: " + str(Average_1))
print("Variance: " + str(Variance_1))
print("Satandard Deviation: " + str(SD_1))

print("\nRubber Width")
print("Average: " + str(Average_2))
print("Variance: " + str(Variance_2))
print("Satandard Deviation: " + str(SD_2))

print("\nPencil Length")
print("Average: " + str(Average_3))
print("Variance: " + str(Variance_3))
print("Satandard Deviation: " + str(SD_3))

print("\nPencil Width")
print("Average: " + str(Average_4))
print("Variance: " + str(Variance_4))
print("Satandard Deviation: " + str(SD_4))
````

หลังจากทำการ plot graph แล้ว ก็สามารถทำ Linear Regression ได้

image.png

*   stats.linregress(x,y) เป็นฟังก์ชั่นในโมดูล scipy.stats ใน Python ที่ใช้สำหรับคำนวณหาสมการเส้นตรง (linear regression) จากข้อมูลที่เรามีอยู่ เพื่อหาค่า intercept (จุดตัดแกน y) และ slope (ความชัน) ของเส้นตรงนั้น
*   Parameters ของฟังก์ชั่น stats.linregress() ประกอบด้วยตัวแปร x และ y ซึ่งเป็น numpy arrays ที่เก็บข้อมูลของตัวแปรต้น (independent variable) และตัวแปรตาม (dependent variable) ตามลำดับ โดยที่ตัวแปร x จะมีความยาวเท่ากับตัวแปร y
*   การทำงานของฟังก์ชั่น stats.linregress() นั้นจะประมวลผลและคำนวณหา intercept และ slope ของเส้นตรงที่เหมาะสมที่สุด โดยใช้วิธี Ordinary Least Squares (OLS) ซึ่งเป็นวิธีที่ใช้คำนวณหาจุดตัดแกน y และความชันของเส้นตรงในการ fitting โมเดล linear regression

*   ฟังก์ชั่น stats.linregress() จะส่งคืนผลลัพธ์เป็น tuple ที่ประกอบด้วย 5 ค่า ได้แก่:

1.   slope: ค่าความชันของเส้นตรงที่ fitting กับข้อมูล
2.   intercept: ค่าจุดตัดแกน y ของเส้นตรงที่ fitting กับข้อมูล
3.   rvalue: ค่า correlation coefficient หรือ coefficient of determination (R-squared) ซึ่งแสดงถึงความสัมพันธ์ระหว่างตัวแปร x และ y
4.   pvalue: ค่า p-value ของการทดสอบสมมติฐานว่าค่า slope ไม่เท่ากับศูนย์
5.   stderr: ค่า standard error ของ slope

````python
import matplotlib
import matplotlib.pyplot as plt
from scipy import stats 

x = [40, 170, 95, 35, 45, 40, 125, 85, 30, 75] # ความยาว (มม.)
y = [20, 9, 7,17, 15, 35, 8, 7, 12, 10]        # ความกว้าง (มม.)

slope, intercept, r, p, std_err = stats.linregress(x, y)

print(slope)
print(intercept)
print(r)
print(p)
print(std_err)

def Linear(x):         # Linear Regression
  return slope * x + intercept

Line = list(map(Linear, x))
print(Line)

plt.scatter(x,y)
plt.plot(x, Line)
plt.show()      
````

Decision Tree เป็นอัลกอริทึมที่นิยมใช้ในการแบ่งกลุ่มข้อมูล (classification) หรือทำนายค่าต่าง ๆ (prediction) โดยอิงจากตัวแปรต้น (input variables) และส่งผลให้ได้ตัวแปรเป้าหมาย (output variable) ที่เป็น categorical หรือ discrete value โดยทำการสร้าง model ใหม่ๆ จากการแบ่งกลุ่มข้อมูลอย่างเป็นลำดับเพื่อทำนายผลลัพธ์ที่เป็นไปได้

* โดยการสร้าง Decision Tree จะเริ่มจากการเลือก feature หรือตัวแปรต้นที่สามารถแยกแยะกลุ่มของ output variable ได้ดีที่สุด แล้วแบ่งกลุ่มข้อมูลตาม feature นั้นๆ ต่อไปจนกว่าจะไม่สามารถแบ่งแยกกลุ่มข้อมูลได้อีก หรือว่างไม่มีข้อมูลเลย โดยแต่ละการแบ่งกลุ่มจะมีค่าของเกณฑ์ชี้วัดความสมดุล (impurity measure) เช่น Gini impurity หรือ Entropy ในการจัดการกลุ่มข้อมูล

* Decision Tree เป็นอัลกอริทึมที่ง่ายต่อการอธิบายและตรวจสอบผลลัพธ์ แต่ยังมีปัญหาเกี่ยวกับ overfitting ได้ง่าย ดังนั้นจึงมีการพัฒนาอัลกอริทึมอื่นๆ เช่น Random Forest, Gradient Boosting และ AdaBoost ที่ใช้ Decision Tree เป็นตัวพื้นฐาน และสามารถลดปัญหา overfitting ได้ดีกว่า Decision Tree โดยเฉพาะในข้อมูลที่ซับซ้อนมาก

Decision Tree เป็นหนึ่งใน Algorithm ที่เอาไว้ใช้สำหรับการทำ Machine Learning เนื่องจากสามารถแสดงแผนผังการคำนวนของ Machine Learning ได้

* **dtree = DecisionTreeClassifier()** คือการสร้าง instance ของคลาส DecisionTreeClassifier ซึ่งเป็นโมเดลสำหรับการเรียนรู้จากข้อมูลที่เป็นต้นไม้การตัดสินใจ (decision tree) ในการจำแนกประเภทของข้อมูล (classification)

* โดยโมเดล Decision Tree จะใช้ข้อมูลในการสร้างโครงสร้างต้นไม้การตัดสินใจ ซึ่งจะแบ่งแยกข้อมูลตามเงื่อนไขที่เกี่ยวข้องกับการจำแนกประเภทของข้อมูล โดยใช้ค่า impurity ของข้อมูลเป็นตัววัดความสำคัญในการสร้างเงื่อนไข โดย impurity จะถูกวัดโดยใช้สูตรต่าง ๆ ได้แก่ Gini index, entropy, หรือ classification error

* โมเดล Decision Tree จะเรียนรู้จากข้อมูลแบบ Supervised Learning โดยมักจะมีการแบ่งข้อมูลเป็น training set และ test set เพื่อสร้างโครงสร้างต้นไม้การตัดสินใจจาก training set และใช้ test set เพื่อวัดประสิทธิภาพของโมเดลที่สร้างขึ้น ซึ่ง Decision Tree นั้นเป็นโมเดลที่เรียนรู้แบบ non-parametric ซึ่งไม่มีค่าพารามิเตอร์ใดๆ ที่ต้องกำหนดมาก่อนใช้งาน

* Gini หรือ Gini Index เป็นตัววัดความไม่สมบูรณ์แบบ (impurity) ของชุดข้อมูล ซึ่งใช้ในการสร้างโมเดลต้นไม้การตัดสินใจ (Decision Tree) ในการจำแนกประเภทของข้อมูล โดยที่ค่า Gini Index มีค่าอยู่ระหว่าง 0 ถึง 1 ยิ่งน้อยยิ่งดี โดยที่ค่า 0 หมายถึงชุดข้อมูลไม่มีความไม่สมบูรณ์แบบ (pure) คือทุกคลาสของข้อมูลมีจำนวนเท่ากัน ส่วนค่า 1 หมายถึงชุดข้อมูลมีความไม่สมบูรณ์แบบสูงสุด คือข้อมูลถูกแบ่งออกเป็นคลาสต่างๆ โดยที่ไม่มีคลาสใดที่ซ้ำกันเลย

* ในการสร้างโมเดล Decision Tree ค่า Gini Index จะถูกนำมาใช้เพื่อวัดความสำคัญของแต่ละเงื่อนไขในการแบ่งข้อมูล โดยการแบ่งข้อมูลที่นำไปสร้างเงื่อนไขที่ให้ค่า Gini Index ต่ำที่สุดจะถือว่าเป็นการแบ่งข้อมูลที่ดีที่สุด

* **dtree.fit(X, y)** คือการสร้างโมเดลต้นไม้การตัดสินใจ (Decision Tree) 
* โดยใช้ข้อมูลใน X เป็นตัวแปรอิสระ (features) และ y เป็นตัวแปรตาม (target) ซึ่งมักจะเป็นค่าที่ต้องการทำนาย หรือเป็นคลาสของข้อมูล โดยโมเดลจะเรียนรู้โดยการแบ่งข้อมูลออกเป็นกลุ่มๆ ตามเงื่อนไขที่เหมาะสม 
* ใช้วิธีการแบ่งข้อมูลตามค่า Gini Index หรือ Entropy และสร้างโครงสร้างต้นไม้การตัดสินใจขึ้นมา 
* โดยโมเดลจะตัดสินใจด้วยการตรวจสอบค่าของตัวแปรอิสระแต่ละตัวว่ามีผลกระทบต่อตัวแปรตามหรือไม่ และจะเรียนรู้ในขั้นตอนนี้ว่าการแบ่งข้อมูลด้วยตัวแปรอิสระตัวไหนจะทำให้ค่า Gini Index หรือ Entropy ลดลงมากที่สุด จนกระทั่งข้อมูลถูกแบ่งออกเป็นกลุ่มของคลาสที่มีค่า Gini Index หรือ Entropy ต่ำสุดแล้ว 
* โดยจะมีการกำหนดเกณฑ์ค่าความลึกของต้นไม้ (max_depth) เพื่อป้องกันการเกิด overfitting ในกรณีที่โมเดลมีขนาดใหญ่เกินไป หลังจากสร้างโมเดลต้นไม้เสร็จสิ้นแล้ว จะสามารถนำโมเดลไปใช้งานสำหรับการทำนายผลได้

````python
import matplotlib
import pandas
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt

# อ่านไฟล์ที่ชื่อว่า myfile.csv
df = pandas.read_csv("myfile.csv")

# Column ที่สำหรับใช้ในการทำ Decision Tree
features = ['Length', 'Width']

X = df[features]
y = df['Item']

dtree = DecisionTreeClassifier()
dtree = dtree.fit(X,y)

tree.plot_tree(dtree, feature_names=features)
````

* ซึ่งในที่นี้คือ [[40, 10]] โดยโมเดลจะใช้ต้นไม้การตัดสินใจเพื่อทำนายผลลัพธ์ว่าข้อมูลดังกล่าวจะอยู่ในกลุ่มคลาสใด โดยจะตรวจสอบเงื่อนไขของตัวแปรอิสระในแต่ละโหนดของต้นไม้ จนกว่าจะถึงโหนดที่เป็นใบ (leaf node) ที่เกี่ยวข้องกับกลุ่มคลาสที่สอดคล้องกับข้อมูลดังกล่าว และคืนค่าผลลัพธ์ที่ได้กลับเป็นคลาสที่โมเดลทำนายได้

* **AgglomerativeClustering** เป็นอัลกอริทึมในกลุ่มของ clustering algorithm ที่ใช้ในการแบ่งกลุ่มของข้อมูล (clustering) โดยที่การแบ่งกลุ่มนี้จะเน้นไปยังการรวมกลุ่มของข้อมูลที่มีความคล้ายคลึงกัน (similarity) ซึ่งใน AgglomerativeClustering นั้นจะเริ่มต้นด้วยการจัดแบ่งแต่ละข้อมูลในกลุ่มของตัวเอง จากนั้นก็จะเริ่มทำการรวมกลุ่มข้อมูลที่มีความคล้ายคลึงกัน โดยจะใช้วิธีการวัดความคล้ายคลึง (similarity metric) เพื่อเปรียบเทียบระหว่างข้อมูลในแต่ละกลุ่ม เมื่อได้กลุ่มที่ต้องการแล้ว จะสร้าง dendrogram ซึ่งเป็นต้นไม้ที่แสดงความสัมพันธ์ของการรวมกลุ่มและข้อมูลที่อยู่ในกลุ่มดังกล่าว จากนั้นจะเลือกตัด dendrogram ที่ต้องการตามความต้องการ จนได้จำนวนกลุ่มที่ต้องการสุดท้าย

* **dendrogram และ linkage** เป็นฟังก์ชันของ Scipy ที่ใช้ในการสร้าง dendrogram ของ hierarchical clustering จาก linkage matrix ที่ได้จากการใช้วิธีการรวมกลุ่มข้อมูล (clustering) โดย linkage matrix เป็นเมทริกซ์ที่แสดงค่าความคล้ายคลึงของแต่ละกลุ่มข้อมูล ซึ่งจะถูกใช้ในการเชื่อมต่อกลุ่มของข้อมูลเข้าด้วยกัน โดยวิธีการเชื่อมต่อกลุ่มข้อมูลจะแบ่งออกเป็น 3 วิธีคือ single linkage, complete linkage, และ average linkage โดยแต่ละวิธีจะมีวิธีการคำนวณความคล้ายคลึงที่แตกต่างกัน

    * **ฟังก์ชัน linkage** จะใช้ในการสร้าง linkage matrix โดยกำหนดข้อมูลที่จะใช้ clustering และ method ของการคำนวณความคล้ายคลึง และคืนค่า linkage matrix ออกมา

    * **ฟังก์ชัน dendrogram** จะใช้ในการสร้าง dendrogram จาก linkage matrix ที่ได้จากฟังก์ชัน linkage โดยกำหนด linkage matrix และ label ของข้อมูลแต่ละตัว และคืนค่า dendrogram ออกมา โดยสามารถกำหนดค่าต่างๆ เพื่อปรับแต่งรูปแบบของ dendrogram ได้ตามต้องการ

* **คำสั่ง data = list(zip(x, y))** จะทำการรวมข้อมูลจาก list x และ y เข้าด้วยกันเป็น tuple โดยเรียงข้อมูลจากตัวแรกถึงตัวสุดท้ายของแต่ละ list แล้วนำมาสร้างเป็น list ใหม่ ตัวอย่างเช่น ถ้า x = [1, 2, 3] และ y = [4, 5, 6] จะได้ผลลัพธ์เป็น [(1, 4), (2, 5), (3, 6)] ซึ่งจะเป็น list ของ tuple ที่มีข้อมูลจาก list x และ y รวมกันเข้าด้วยกันเป็นคู่ลำดับ

* **คำสั่ง hierarchical_cluster = AgglomerativeClustering(n_clusters=Cluster_Count, affinity='euclidean', linkage='ward')** จะสร้าง object ที่เป็น hierarchical clustering model โดยกำหนดจำนวน clusters ที่ต้องการ (Cluster_Count) และกำหนดวิธีการคำนวณความคล้ายคลึงระหว่างข้อมูล (affinity) และวิธีการเชื่อมต่อกลุ่มข้อมูล (linkage) โดยในที่นี้ใช้ 'euclidean' เพื่อคำนวณความคล้ายคลึงระหว่างข้อมูลโดยใช้ Euclidean distance และ 'ward' เพื่อเชื่อมต่อกลุ่มข้อมูลโดยใช้วิธีการ Ward's minimum variance method

* **คำสั่ง labels = hierarchical_cluster.fit_predict(data)** จะใช้ hierarchical clustering model ที่สร้างขึ้นก่อนหน้านี้ (hierarchical_cluster) ในการทำ hierarchical clustering บนข้อมูล (data) และคืนค่า label ของแต่ละ cluster ออกมา โดย label จะเป็นเลขจำนวนเต็มเริ่มจาก 0 แทน cluster ต่างๆ ในผลลัพธ์ของการ clustering

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage

val = pd.read_csv('myfile.csv')
x = val['Length'].tolist()
y = val['Width'].tolist()

print("Length value {}".format(x))
print("Width value {}".format(y))

# Clustering Algorithm
data = list(zip(x, y))

# จำนวนกานแบ่งกลุ่ม
Cluster_Count = 2

hierarchical_cluster = AgglomerativeClustering(n_clusters=Cluster_Count, affinity='euclidean', linkage='ward') 
labels = hierarchical_cluster.fit_predict(data)

plt.scatter(x, y, c=labels)
plt.show()
````

````python
import matplotlib
import pandas
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt

# อ่านไฟล์ที่ชื่อว่า iris.csv
df = pandas.read_csv("iris.csv")
# Column ที่สำหรับใช้ในการทำ decision tree
features = ['sepal.length', 'sepal.width','petal.length','petal.width']

X = df[features]
y = df['variety'] 					# ผลลัพธ์ที่เอาไว้ใช้ในการทำนาย

dtree = DecisionTreeClassifier()
dtree = dtree.fit(X, y)

tree.plot_tree(dtree, feature_names=features)

plt.rcParams['figure.figsize'] = (30, 6)      # ปรับขนาดภาพ
plt.savefig("abc.png")                        # บันทึกภาพ

# สำหรับ predict
print(dtree.predict([[5,2.1,1,.1]]))      	# sepal.length, sepal.width, petal.length, petal.width
print(dtree.predict([[7,3.2,4.7,1.4]]))		  # sepal.length, sepal.width, petal.length, petal.width
````

Clustering (Unsupervised Learning)

````python
# Code here
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage

val = pd.read_csv('iris.csv')
x = val['sepal.length'].tolist()
y = val['sepal.width'].tolist()

a = val['petal.length'].tolist()
b = val['petal.width'].tolist()

print("Length value {}".format(x))
print("Width value {}".format(y))

# Clustering algorithm
data = list(zip(x, y))
data2 = list(zip(a, b))

# จำนวนการแบ่งกลุ่ม
Cluster_Count = 3

hierarchical_cluster = AgglomerativeClustering(n_clusters=Cluster_Count, affinity='euclidean', linkage='ward')
labels = hierarchical_cluster.fit_predict(data)

plt.scatter(x, y, c=labels)
plt.show()
````

````python
hierarchical_cluster_2 = AgglomerativeClustering(n_clusters=Cluster_Count, affinity='euclidean', linkage='ward')
labels_2 = hierarchical_cluster_2.fit_predict(data2)

plt.scatter(a, b, c=labels_2)
plt.show()
````
