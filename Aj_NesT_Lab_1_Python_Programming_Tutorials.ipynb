## 5. Python Loops

**Exercise 1.12: Write a program to accomplish this using a for loop.**

Here are the steps to follow:

1. Create a list of words (e.g., ["apple", "banana", "cat", "dog", "elephant", "frog"]).
2. Initialize an empty list new_list to hold the words with more than three characters.
3. Iterate over the list of words using a for loop.
4. Inside the loop, check if the length of the current word is greater than three. If it is, append the word to the new_list variable.
5. After the loop finishes, print the new_list variable.

Screen Shot 2566-03-05 at 01.20.47.png

**Exercise 1.12: Solution**

```
"""
Aj. NesT the Series
5. Python Loops
Exercise 1.12
"""
# Create a list of words
words = ["apple", "banana", "cat", "dog", "elephant", "frog"]

# Initialize an empty list to hold the words with more than three characters
new_list = []

# Iterate over the list of words using a for loop
for word in words:
    # Check if the length of the current word is greater than three
    if len(word) > 3:
        # If the length of the current word is greater than three, append it to the new list
        new_list.append(word)

# Print the new list
print("Words with more than three characters:", new_list)
```

**Exercise 1.13: Write a power calculator program using a while loop.**

Screen Shot 2566-03-05 at 01.21.22.png

**Exercise 1.13: Solution**

```
"""
Aj. NesT the Series
5. Python Loops
Exercise 1.13
"""
# Prompt the user for input
base = int(input("Enter the base: "))
exponent = int(input("Enter the exponent: "))

# Initialize the result and the counter
result = 1
counter = 0

# Use a while loop to calculate the power of the base to the exponent
while counter < exponent:
    result *= base
    counter += 1

# Print the result
print(base, "^", exponent, "=", result)
```

**Exercise 1.17: Write a program creates two sets of integers called set1 and set2, prints the sets, performs various set operations such as union, intersection, difference, and symmetric difference, and then prints the results of these operations.**

Screen Shot 2566-03-05 at 02.18.58.png

**Exercise 1.17: Solution**

```
"""
Aj. NesT the Series
6. Python Data Structures (Sets)
Exercise 1.17
"""
# create two sets of integers
set1 = {1, 2, 3, 4, 5}
set2 = {3, 4, 5, 6, 7}

# print the sets
print("Set 1:", set1)
print("Set 2:", set2)

# perform union operation on the sets
union = set1.union(set2)

# print the union of the sets
print("Union:", union)

# perform intersection operation on the sets
intersection = set1.intersection(set2)

# print the intersection of the sets
print("Intersection:", intersection)

# perform difference operation on the sets
difference = set1.difference(set2)

# print the difference of the sets
print("Difference (set1 - set2):", difference)

# perform symmetric difference operation on the sets
symmetric_difference = set1.symmetric_difference(set2)

# print the symmetric difference of the sets
print("Symmetric difference:", symmetric_difference)
```

## 7. Python Functions

**Exercise 1.19: Write a program that defines global and local variables and then calls the user-defined functions using those variables.**

Screen Shot 2566-03-05 at 02.41.41.png

**Exercise 1.19: Solution**

```
"""
Aj. NesT the Series
7. Python Functions
Exercise 1.19
"""
# Define the global variable
global_var = 5

# Define a function that uses a local variable
def my_function(local_var):
    print("Global variable: ", global_var)
    print("Local variable: ", local_var)

# Define another function that modifies the global variable
def modify_global_var():
    global global_var
    global_var += 1

# Call the functions using global and local variables
local_var = 10
my_function(local_var)        # Output: Global variable: 5, Local variable: 10

modify_global_var()
my_function(local_var)        # Output: Global variable: 6, Local variable: 10
```

**Exercise 1.20: Write a program that calls the user-defined functions.**

Screen Shot 2566-03-05 at 02.46.04.png

**Exercise 1.20: Solution**

```
"""
Aj. NesT the Series
7. Python Functions
Exercise 1.20
"""
# Define the functions
# Function 1
def find_max(num1, num2, num3):
    max_num = num1
    if num2 > max_num:
        max_num = num2
    if num3 > max_num:
        max_num = num3
    return max_num

# Function 2
def is_prime(num):
    if num <= 1:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

# Function 3
def reverse_string(str):
    return str[::-1]

# Function 4
def factorial(num):
    if num == 0:
        return 1
    else:
        return num * factorial(num-1)

# Function 5
def calculate_sum(num_list):
    sum = 0
    for num in num_list:
        sum += num
    return sum

# Call the functions
print(find_max(5, 8, 3))            # Output: 8
print(is_prime(17))                 # Output: True
print(reverse_string("hello"))      # Output: "olleh"
print(factorial(5))                 # Output: 120
print(calculate_sum([2, 3, 5, 7]))  # Output: 17
```
