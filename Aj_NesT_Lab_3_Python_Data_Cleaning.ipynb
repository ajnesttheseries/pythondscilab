**STEP 2:** นำเข้า pandas และ numpy libraries และทำการอ่านไฟล์ .csv ด้วย pandas เก็บเป็น Dataframe

**=> ใช้คำสั่ง .read_csv()** สำหรับอ่านข้อมูลของไฟล์ .csv

**STEP 3:** ตั้งค่าและแสดง index และ size ของ nls97 data

**=> ใช้คำสั่ง set_index()** สำหรับตั้งค่า Index ของ Columns

**STEP 4:** แสดงจำนวน data types และ non-null value

```
"""
Aj. NesT the Series
2. การเลือกและจัดระเบียบคอลัมน์ (Selecting and Organizing Columns)
Exercise 3.2
"""
# Step 1 - 4:
import pandas as pd
import numpy as np
nls97 = pd.read_csv("nls97.csv")
nls97.set_index("personid", inplace=True)
print(nls97.info())
```

**STEP 5:** แปลง data types ของ columns ที่เป็น object ให้เป็น category เพื่อทำข้อมูลให้เหมาะสม

data typs --> category เหมาะสำหรับที่เป็นหมวดหมู่ ข้อมูลที่มีซ้ำ ๆ กัน ที่ไม่ค่อยต่างกันมาก ช่วยในการเรื่องการประหยัด Memory

**=> ใช้คำสั่ง .loc[ ]** ทำการ Filter เข้าถึงข้อมูล -->  คำสั่ง pandas หมวด Filter จะมีทั้ง .loc [  ], .iloc [  ], .query [  ]

**=> ใช้คำสั่ง .select_dtypes** ทำการเลือก data types

**=>ใช้คำสั่ง .apply(lamba function)** เพื่อแปลงข้อมูล

**=> ใช้คำสั่ง .astype** ทำการแปลง data types

```
# Step 5: 
nls97.loc[:, nls97.dtypes == 'object'] = \
  nls97.select_dtypes(['object']).\
  apply(lambda x: x.astype('category'))
print(nls97.info())
# เครื่องหมาย \ คือ เพื่อเชื่อมประโยคให้ต่อกันในอีกบรรทัดนึง กรณีเขียนโปรแกรมแล้วบรรทัดยาว ต้องการขึ้นบรรทัดใหม่
# Syntax --> .loc['row_name' , 'column_name'] สนใจชื่อ
# Syntax --> .iloc['row_index_number' , 'column_index_number'] สนใจตำแหน่ง index ของ row และ column
# loc[: Filter ชื่อทุกตัว, ที่มีเงื่อนไขเป็น 'object'] และทำการเลือก loc ย่อมาจาก location based
# ทำการแปลง colums ที่เป็น 'object' --> 'category'
# ผลปรากฎว่า 'category' ช่วยลด memory (2.7 MB) ได้มากกว่า 'object' (6.1+ MB)
```

**STEP 7:** วิธี Filter ข้อมูล เลือก rows และ columns ที่ต้องการ โดยใช้ [ ] index oprator, .loc, .iloc ของ pandas

![alt text](https://miro.medium.com/max/2000/1*1C-2_cMi-PUz1Tg5Cscs7A.png)

**=> ใช้คำสั่ง .loc[ row_name, column_name ]** 

เลือกเข้าถึง rows แบบใช้ชื่อ 'row_name' และ column แบบใช้ชื่อ 'column_name'

```
# Step 7.1: ใช้คำสั่ง .loc แบบเข้าถึงทุก row (:) และเข้าถึงเฉพาะ column ['gender']     
# .loc['row_name' , 'column_name']  
# .loc[     :     ,    [ ]       ]  => (:) slicing
analysisdemo = nls97.loc[ : , ['gender']]  
print(type(analysisdemo))                   
print(analysisdemo)
```

```
# Step 7.2: ใช้คำสั่ง .loc แบบเข้าถึงเฉพาะ row (start : end) และเข้าถึงเฉพาะ column ['gender']
# .loc[.      'row_name'    , 'column_name']    
# .loc[ row_start : row_end ,      [ ]     ]  => (start : end) slicing
analysisdemo = nls97.loc[100139:100583, ['gender']]  
print(type(analysisdemo))                             
print(analysisdemo)
```

```
# Step 7.3: ใช้คำสั่ง .loc แบบเข้าถึงเฉพาะ row (start : end) และเข้าถึงเฉพาะ column [column1, column2, ...]ที่ต้องการ
# .loc['row_name' ,             'column_name'         ] 
# .loc[     :     , [ col_name1, col_name2, col_name3 ]] => (:) slicing
analysisdemo = nls97.loc[ : , ['gender','birthmonth','birthyear']] 
print(type(analysisdemo))                                          
print(analysisdemo)
```

=> **ใช้คำสั่ง .iloc[row_index_number, column_index_name]** 

เลือกเข้าถึงทุก rows และเลือก column แบบใช้เลขตำแหน่ง index ของ column ที่ต้องการ

```
nls97
```

```
# Step 7.4: ใช้คำสั่ง .iloc เลือกเข้าถึงทุก rows และเลือก column แบบใช้เลขตำแหน่ง index ของ column ที่ต้องการ
# .iloc['row_index_number' , 'column_index_number'] 
# .iloc[        :          ,      [ col_index ]   ]  => (:) slicing, index 0 คือ gender
analysisdemo = nls97.iloc[:, [0]]  
print(type(analysisdemo))         
print(analysisdemo)
#ทดลองเปลี่ยนเลขตำแหน่ง index เพื่อเข้าถึง columns ต่าง ๆ
```

```
# Step 7.5: ใช้คำสั่ง .iloc เลือกเข้าถึง rows และเลือก column แบบใช้เลขตำแหน่ง index ของ column ที่ต้องการ
# .iloc[           'row_index_number'    ,         'column_index_number'          ]   
# .iloc[ row_index_start : row_index_end , [ col_index1, col_index2, col_index3 ] ]  => (start : end) slicing
analysisdemo = nls97.iloc[2:100, [0,1,2]]  
print(type(analysisdemo))                  
print(analysisdemo)
#ทดลองเปลี่ยนเลขตำแหน่ง index เพื่อเข้าถึง row ต่าง ๆ
```
